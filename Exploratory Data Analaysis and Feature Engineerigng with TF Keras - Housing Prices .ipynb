```python
# when passing X and y it is .values is important as tensorflow cannot compute a df or series that well so this puts it back into arraty form
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
# now testtrainsplit 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler  
```

```python
scaler = MinMaxScaler()
```

```python
# scaling and transforming the train set 
X_train = scaler.fit_transform(X_train)
```

```python
# Now doing the same for the test set but instead only trasnform dont fit 
X_test = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
```

```python
# OBSERVING THE SHAPE OF TRAIN DATA 
X_train.shape
# since 19 features - good start for 19 neurons in layers 
```

```python
model = Sequential()

model.add(Dense(19,activation ='relu'))
model.add(Dense(19,activation ='relu'))
model.add(Dense(19,activation ='relu'))
model.add(Dense(19,activation ='relu'))

model.add(Dense(1))
# compiling model with an adam optmiser and mean squared error loss 
model.compile(optimizer='adam', loss= 'mse')
```

```python
model.fit(x=X_train, y=y_train,
          validation_data =(X_test,y_test),
         batch_size = 128, epochs=400)
```

```python
# explore the model history - now we can convert this to a df 
#cna compre oss on training and loss on valiadtion 
pd.DataFrame(model.history.history)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
# here trian loss and val loss both decrease - does not show signs of overfitting 
# CAN COMPARE behaviour or loss and val loss, both lines are the same v close showing it could continue
```

```python
# now for evaluation on test data 
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
# if values are too large and hard to interprete
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
# to evaluate the MAE AND MSE we need to look at the dataframe as a whole 
df['price'].describe()
# if we compare the mean value for the price is 540296.6 the MAE is about a 5th of that which is not a good indicatpr 
```

```python
# we can also use this metric - Best possible score is 1.0, lower values are worse.
explained_variance_score(y_test, predictions )
```

```python
# we can see that for the really expensive hosues the prediction is bad , so might be worht rertrianig model based on lower priced houses 
plt.figure(figsize=(12,6))
plt.scatter(y_test, predictions)
plt.plot(y_test,y_test, 'r')
```
